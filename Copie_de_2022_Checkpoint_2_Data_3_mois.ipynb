## 3.1 API adresses
Le jeu de données suivant recense une sélection des meilleurs restaurants de Paris, à prix très abordable (moins de 15 euros le menu en moyenne).  
En tant que Data Analyst en herbe, tu vas utiliser l'API [suivante](https://adresse.data.gouv.fr/api-doc/adresse), pour récupérer les coordonnées de ces restaurants.

```
import requests
```

```
food_paris = pd.read_csv("https://raw.githubusercontent.com/Gierko/food_paris_under_15_euros/main/food.csv")
```

```
food_paris
```

```
#link_main = '"https://api-adresse.data.gouv.fr/search/?q=8+bd+du+port"'
#address = '54 Via Pietro Mascagni, Catania, Italy'
#ink_end = '&format=json&limit=1'

#link = link_main + address.replace(# It's your turn here. Your goal is to get a link in the right format
#link = link_main + address.replace(' ','+')
#print(link + link_end)
```

```
def API_adresse1(adresse_postale):
  link=f"https://api-adresse.data.gouv.fr/search/?q={adresse_postale}"
  r = requests.get(link).json()
  r1=r['features'][0]['geometry']['coordinates'][::-1]
  return r1
food_paris['corr']=food_paris['adresse'].apply(API_adresse1)
food_paris
```

```
food_paris[['lat','lon']]= food_paris['corr'].to_list()
food_paris
food_paris[['lat','lon']]=food_paris[['lat','lon']].astype(float)
```


## 3.2  Folium
Tu vas maintenant projeter ces restaurants sur une carte, en utilisant la librairie `Folium`.


```
import folium
```

```

map = folium.Map(location=[food_paris['lat'].mean(), food_paris['lon'].mean()],
                 zoom_start=15,
                 control_scale=True)

for index, row in food_paris.iterrows():
    folium.Marker([row['lat'], row['lon']], popup = row['nom']).add_to(map)

map
```
